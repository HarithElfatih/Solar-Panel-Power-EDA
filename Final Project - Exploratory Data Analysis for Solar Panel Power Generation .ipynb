## <font color = Darkblue> Solar Power Analysis By Seasons </font>

```python
# Here the weather conditions data has been orgnized in Four seperate groups to visulize the difference in the weather 
#  conditions and the correspondence output ower for every season

# Grouping Winter season data in one dataFrame (December, January, February)

December = data[data.index.month == 12] 
January = data[data.index.month == 1]
February = data[data.index.month == 2]

winter_months = pd.concat([December, January, February], axis=0)


# Grouping Spring season data in one dataFrame (March, April, May)

March = data[data.index.month == 3] 
April = data[data.index.month == 4]
May   =  data[data.index.month == 5]

spring_months = pd.concat([March , April, May], axis=0)

# Grouping Summer season data in one dataFrame (June, July, August)

June = data[data.index.month == 6]
July = data[data.index.month == 7]
August = data[data.index.month == 8]

summer_months = pd.concat([June , July, August], axis=0)

# Grouping Fall season data in one dataFrame (September, October, November)

Septemebr = data[data.index.month == 9]
October = data[data.index.month == 10]
November = data[data.index.month == 11]

fall_months = pd.concat([Septemebr , October, November], axis=0)

```

## Visualization of the The Data By Seasons
